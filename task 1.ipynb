```python
import os
import random
import matplotlib.gridspec as gs
import cv2 as cv2
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def openAsImage(fname):        
    fits_image = fits.open(fname)
    #print(fits_image.info())
    return fits_image[0].data
```

```python
def produceImage(name, ax, title, cmap_name = 'CMRmap'):
    mm = openAsImage(name)
    mm = mm.astype(np.float32)

    #mm = np.log(mm)
    ax.imshow((mm), cmap=cmap_name)
    ax.set_title(title, fontsize=18)
    ax.tick_params(labelbottom=False, labelleft=False)
```

```python
def drawImages(name, cmap = 'CMRmap'):
    name_mm = name
    name_dm = name[:-6] + "2.fits"
    titles = ['model_map', 'deconvolved_map']
    images = [name_mm, name_dm]

    i = 0
    fig, axs = plt.subplots(figsize=(20,20),nrows=1, ncols=2)
    for ax in axs.flatten():
        produceImage(images[i], ax, titles[i], cmap)
        i = i + 1
    plt.suptitle(name[name.find('_')+1:-15], fontsize = 30)        
    plt.show()
```

```python
cmaps = [
            'flag', 'prism', 'ocean', 'gist_earth', 'terrain', 'gist_stern',
            'gnuplot', 'gnuplot2', 'CMRmap', 'cubehelix', 'brg',
            'gist_rainbow', 'rainbow', 'jet', 'nipy_spectral', 'gist_ncar']
```

```python
names = []
s = "./data/"
for i in os.listdir(s):
    if (i[-6:] == "1.fits"):
        names.append(s + i)
```

```python
 for i in names[0:5]:
    drawImages(i)
    print "-----------------------------------------------------------------------------------------------------"
```

```python
for name in names[0:5]:
    for cmap in cmaps:
        print 'File = %s, Cmap = %s'%(name,cmap)
        drawImages(name, cmap)
        print '-------------------------------------------------------------------------------------'
```
